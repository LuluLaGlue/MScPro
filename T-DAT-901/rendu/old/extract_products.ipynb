# Extract Products From KaDo.csv

```python
import pandas as pd
import os
```

```python
df = pd.read_csv("..{}data{}KaDo.csv".format(os.sep, os.sep), sep=",")

df.head()
```

```python
df = df[["FAMILLE", "MAILLE", "UNIVERS", "LIBELLE", "PRIX_NET"]]

df.head()
```

```python
df.shape[0]
```

```python
df_product = df.drop_duplicates(subset="LIBELLE").sort_values(["FAMILLE", "MAILLE", "UNIVERS"]).reset_index().drop(["index"], axis=1)

df_product.head(10)
```

```python
df_product.shape[0]
```

```python
df_product = df_product.sample(frac=1)
df_product.to_csv("..{}data{}KaDo_Products.csv".format(os.sep, os.sep), sep=",")
```
